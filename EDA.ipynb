```python
import os
import warnings
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from IPython.core.interactiveshell import InteractiveShell
```

```python
plt.rc('font', family='GULIM')
warnings.filterwarnings(action='ignore')
InteractiveShell.ast_node_interactivity = "all"
```

# Data Check

```python
input_dir = 'Database/'
file_data = 'train.csv'
data = pd.read_csv(os.path.join(input_dir, file_data))
data=data.drop(columns= ['ID'])
train, test = train_test_split(data, test_size=0.2, random_state=42, shuffle=True)
```

```python
discrete_list=['대출기간', '근로기간', '주택소유상태', '대출목적']
continuous_list=['대출금액', '연간소득', '총상환원금', '총상환이자', '총연체금액', '부채_대비_소득_비율', '최근_2년간_연체_횟수', '총계좌수', '연체계좌수']
target='대출등급'
```

```python
# Create a figure with individual subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(16, 10))

# Plot histograms for each numerical column
for i, column in enumerate(continuous_list):
    row, col = i // 3, i % 3
    axes[row, col].hist(data[column], bins=10)
    axes[row, col].set_title(column)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 연간소득의 분포를 확인하는 KDE 플롯
plt.figure(figsize=(10, 6))
sns.kdeplot(train['총상환이자'], shade=True)
plt.title('연간소득의 분포')
plt.xlabel('연간소득')
plt.ylabel('밀도')
plt.show()
```

```python
# Create a figure with individual subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16, 10))

# Plot boxplots for each numerical column
for i, column in enumerate(continuous_list):
    row, col = i // 3, i % 3
    sns.boxplot(x='대출등급', y=column, data=data, ax=axes[row, col])
    axes[row, col].set_title(column)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
class_counts = data['대출등급'].value_counts()
plt.pie(class_counts, labels=class_counts.index, startangle=140, autopct='%1.1f%%')
plt.axis('equal')
plt.title('대출등급')
plt.show()
```

```python
print('범주 갯수')
for i, name in enumerate(discrete_list):
    train_class = train[name].value_counts()
    test_class = test[name].value_counts()
    print(f'{name}: {len(set(train_class.index))} 개')
    print(f'{name}: {len(set(test_class.index))} 개')

    not_in_train = [item for item in test_class.index if item not in train_class.index]
    test_only_counts = test[name].isin(not_in_train).sum()
    test_only_ratio = test_only_counts / test[name].value_counts().sum()
    
    if name != 'customer_country':
        print(set(train_class.index))
        print(set(test_class.index))

    if not_in_train:
        print(f'{name}: Train에 없는 Test 범주갯수: {len(not_in_train)}개')

    else:
        print(f'{name}: 모든 Test 범주가 Train에 존재함.')
    print(f"Only Test 범주의 비율: {test_only_ratio * 100:.2f}%")
    print('')
```

```python
# 서브플롯 생성
fig, axes = plt.subplots(nrows=len(discrete_list), ncols=1, figsize=(10, 5 * len(discrete_list)))
target_column='대출등급'
# 각 이산형 변수에 대해 타겟 변수의 분포를 시각화
for i, col in enumerate(discrete_list):
    sns.countplot(x=col, hue=target_column, data=train, ax=axes[i])
    axes[i].set_title(f'{col}에 따른 {target_column}의 분포')
    axes[i].legend(title=target_column, loc='upper right')

# 레이아웃 조정
plt.tight_layout()

# 그래프 보여주기
plt.show()
```
